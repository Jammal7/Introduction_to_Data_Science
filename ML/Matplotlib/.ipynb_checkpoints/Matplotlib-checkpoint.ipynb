```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 10)
y = 2 * x
```

```python
x 
```

```python
y
```

```python
plt.plot(x,y)
# Add title
plt.title("String Title");
# Add labels
plt.xlabel("x-axis")
plt.ylabel("y-axis")
# Add limit for the point
plt.xlim(0,9)
plt.ylim(0,18)
# Save the graph
# plt.savefig("myfirstplot.jpeg")
```

## Figures and Axes

```python
a = np.linspace(0,10,11)
b = a**4
```

```python
fig = plt.figure()

# Large axes
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)

axes1.set_xlim(0,8)
axes1.set_ylim(0,8000)
axes1.set_xlabel('A')
axes1.set_ylabel('B')
axes1.set_title('Power of 4')
# Small axes
axes2 = fig.add_axes([0.2,0.5,0.25,0.25])
axes2.plot(a,b)

axes2.set_xlim(1,2)
axes2.set_ylim(0,50)
axes2.set_xlabel('A')
axes2.set_ylabel('B')
axes2.set_title('Zoomed In')

plt.show()
```
