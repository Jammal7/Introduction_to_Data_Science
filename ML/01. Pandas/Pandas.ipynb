## Conditional Filtering

It allows to select **rows** based of a condition on a column.

```python
df = pd.read_csv('Data/tips.csv')
df.head()
```

### Filter based on a single condition

```python
df[df['total_bill'] > 40]
```

```python
df[df['sex'] == "Male"]
```

```python
df[df['size'] >= 3]
```

### Filter based on a Multiple condition

1. AND & --- BOTH CONDITION NEED TO BE TRUE
2. OR | --- EITHER CONDITION IS TRUE

```python
df[(df['total_bill'] > 30) & (df['sex'] == "Male")]
```

```python
df[(df['total_bill'] >= 30) | (df['time'] == "Lunch")]
```

```python
df[(df['day'] == 'Sun') | (df['day'] == "Sat") | (df['day'] == "Fri")]
```

The **OR** COND can get really big and difficult if we have multiple conditions, the best way to solve something like this is to use the **__isin__** condition

```python
options = ['Sat', 'Sun']
df[df['day'].isin(options)]
```

## Apply Method

```python
# Single column
def last_four(num):
    return str(num)[-4:]
```

```python
df['last_four'] = df['CC Number'].apply(last_four)
df.head()
```

```python
def yelp(price):
    if price < 10:
        return ''
    elif price >= 10 and price < 30:
        return ''
    else:
        return ''
```

```python
df['yelp'] = df['total_bill'].apply(yelp)
df.head()
```

```python
# Multiple Columns
def quality(total_bill, tip):
    if tip/total_bill > 0.25:
        return "Generous"
    else:
        return "Other"
```

```python
df['Quality'] = df[['total_bill', 'tip']].apply(lambda df: quality(df['total_bill'], df['tip']), axis=1)
df.head()
```

```python
df['Qual'] = np.vectorize(quality)(df.total_bill, df.tip)
df.head()
```

## Merge

```python
registrations = pd.DataFrame({'reg_id':[1,2,3,4],'name':['Andrew','Bobo','Claire','David']})
logins = pd.DataFrame({'log_id':[1,2,3,4],'name':['Xavier','Andrew','Yolanda','Bobo']})
```

```python
logins
```

```python
registrations
```

```python
# help(pd.merge)
```

### Inner Join

**Match up where the key is present in BOTH tables. There should be no NaNs due to the join, since by definition to be part of the Inner Join they need info in both tables.**
**Only Andrew and Bobo both registered and logged in.**

```python
pd.merge(registrations, logins, how='inner', on='name')
```

### Left Join

**Match up AND include all rows from Left Table.**
**Show everyone who registered on Left Table, if they don't have login info, then fill with NaN.**

```python
pd.merge(left=registrations, right=logins, how='left', on='name')
```

### Right Join

**Match up AND include all rows from Right Table.**
**Show everyone who logged in on the Right Table, if they don't have registration info, then fill with NaN.**

```python
pd.merge(left=registrations, right=logins, how='right', on='name')
```

### Outer Join

**Match up on all info found in either Left or Right Table.**
**Show everyone that's in the Log in table and the registrations table. Fill any missing info with NaN**

```python
pd.merge(left=registrations, right=logins, how='outer', on='name')
```

## Join on Index or Column

**Use combinations of left_on,right_on,left_index,right_index to merge a column or index on each other**

```python
registrations = registrations.set_index('name')
registrations
```

```python
pd.merge(registrations, logins, left_index=True, right_on='name', how='inner')
```
